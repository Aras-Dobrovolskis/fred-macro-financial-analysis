# Finance Project: FRED Monthly Macro-Financial Analysis

This notebook loads the cleaned monthly dataset and SQLite database, then visualizes the relationships between interest rates, inflation, volatility, and the S&P 500.

```python
from pathlib import Path
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent
DATA_PATH = ROOT / 'output' / 'fred_monthly_merged.csv'
DB_PATH = ROOT / 'output' / 'finance_project.db'
SQL_PATH = ROOT / 'sql' / 'analysis_queries.sql'

print('ROOT =', ROOT)
print('DATA_PATH =', DATA_PATH)
print('DB_PATH =', DB_PATH)
print('SQL_PATH =', SQL_PATH)
```

```python
df = pd.read_csv(DATA_PATH, parse_dates=['date'])
df = df.sort_values('date')
df.head(10)
```

```python
df.info()
```

```python
with sqlite3.connect(DB_PATH) as conn:
    query = 'SELECT date, sp500_pct_change, inflation_pct_change, rate_regime, volatility_regime FROM fred_monthly_data ORDER BY date DESC LIMIT 10'
    sample = pd.read_sql_query(query, conn, parse_dates=['date'])
sample
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df['date'], df['sp500'], color='tab:blue', linewidth=2)
ax.set_title('S&P 500 Monthly Close')
ax.set_xlabel('Date')
ax.set_ylabel('S&P 500')
fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df['date'], df['fedfunds'], color='tab:red', linewidth=2)
ax.set_title('Fed Funds Rate')
ax.set_xlabel('Date')
ax.set_ylabel('Fed Funds Rate')
fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(df['fedfunds'], df['sp500_pct_change'], color='tab:purple', alpha=0.7, edgecolor='black')
ax.set_title('S&P 500 Monthly Returns vs Fed Funds Rate')
ax.set_xlabel('Fed Funds Rate')
ax.set_ylabel('S&P 500 Monthly Return')
fig.tight_layout()
plt.show()
```
